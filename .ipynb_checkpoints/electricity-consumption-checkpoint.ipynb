```python
df=train_set.copy()
df_label=df["Date"].copy()
df=df.drop("Date",axis=1)
# df.describe()
```

```python
# feature scaling

# df.describe()
from sklearn.preprocessing import StandardScaler
feature_scale=StandardScaler()
data_num_scaled_tr=pd.DataFrame(feature_scale.fit_transform(df.values),columns=df.columns)
data_num_scaled_tr.head()
```

```python
# import datetime
# date_features = df_label
# df_label = pd.to_datetime(df_label)

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

final = pd.concat([data_num_scaled_tr,df_label], axis=1)

param_grid = [{'n_estimators': [3, 4, 6, 10, 30], 'max_features': [2, 6, 8, 15]}]
forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(final, df_label)

print("Best Parameters: ", grid_search.best_params_)
print("Best Estimator: ", grid_search.best_estimator_)
```
